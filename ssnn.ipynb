```python
def calculate(a,b,symbol_list):
    result = []
    for i in range(len(symbol_list)):
        if symbol_list[i] == '+':
            result.append(a[i]+b[i])
        if symbol_list[i] == '-':
            result.append(a[i]-b[i])
        if symbol_list[i] == '/':
            result.append(a[i]/b[i])
        if symbol_list[i] == '*':
            result.append(a[i]*b[i])
    return result
```

```python
from random import seed,randint
```

```python
seed(0)
max_val=1000

a = [randint(0,20) for x in range(max_val)]
b = [randint(0,20) for x in range(max_val)]
#symbols = ['+','-','/','*']
symbols = ['+','-','*']
symbol_list = [symbols[randint(0,2)] for i in range(max_val)]

result = calculate(a,b,symbol_list)
#print(a[:10],b[:10],symbol_list[:10],result[:10],sep='\n')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data=zip(a,b,symbol_list,result),columns=['a','b','symbol','result'])
```

```python
#df.head(6)
```

```python
X = pd.get_dummies(df['symbol'], columns=['type']).iloc[:,:-1]
X = pd.DataFrame(zip(df['a'],df['b'],X['*'],X['+']),columns=['a','b','*','+'])
y = df.iloc[:,-1:]
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
```

```python
model = Sequential()
model.add(Dense(2, input_dim=4, activation='relu'))
model.add(Dense(1,activation='relu'))
# Compile model
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.summary()
```

```python
es = EarlyStopping(monitor='loss', mode='min', verbose=1, patience=10)
```

```python
history = model.fit(X_train, y_train, epochs=50000, batch_size=1, validation_data=(X_test,y_test), verbose=1, callbacks=es)
```

```python
from matplotlib import pyplot
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```
